### sigmoid 함수
- 시그모이드 함수는 S자형 곡선 또는 시그모이드 곡선을 갖는 수학 함수
- ${\displaystyle S(x)={\frac {1}{1+e^{-x}}}={\frac {e^{x}}{e^{x}+1}}.}$
- x가 큰 값을 가지면 s(x)는 1에 가까워지고, x가 작은 값을 가지면 s(x)는 0에 가까워진다
- 0 또는 1, 두 개의 값중 하나를 고를 때 유용하게 쓰인다

```python
# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# 시그모이드 시각화
for x in np.linspace(-10, 10, 50):
    plt.plot(x, sigmoid(x), 'ro')
plt.show()
```
